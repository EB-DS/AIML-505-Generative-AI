## 2.2 Generative Artificial Intelligence

Generative AI models are typically trained on massive datasets using self-supervised learning techniques, enabling them to capture complex linguistic, visual, and semantic relationships without requiring manually labeled data. ### 2.3 Large Language Models (LLMs)

Large Language Models (LLMs) are a class of deep learning models specifically designed to understand, generate, and manipulate human language. Built upon transformer architectures, LLMs are trained on massive corpora of text using self-supervised learning techniques that enable them to predict the next token in a sequence. Through this training process, LLMs learn complex linguistic structures, semantic relationships, contextual dependencies, and patterns of human language without requiring manually labeled datasets (Brown et al., 2020; Devlin et al., 2019; Vaswani et al., 2017).

Modern LLMs contain billions or even trillions of parameters that capture statistical representations of language. Rather than memorizing information, these models learn mathematical representations that allow them to generate coherent, contextually appropriate, and often highly informative responses. Consequently, LLMs have become the foundation for numerous generative AI applications, including conversational assistants, intelligent search, document summarization, software development, translation, question answering, and knowledge management.

### 2.4 Transformer Architecture

The Transformer architecture is the foundational deep learning framework that powers modern Large Language Models (LLMs). Introduced by Vaswani et al. (2017), the Transformer fundamentally changed natural language processing by replacing sequential processing methods, such as Recurrent Neural Networks (RNNs) and Long Short-Term Memory (LSTM) networks, with a mechanism known as **self-attention**. This innovation enables models to process entire sequences of text simultaneously while learning the contextual relationships between words regardless of their position within a sentence.

Unlike earlier neural network architectures that processed language one word at a time, Transformer models analyze all tokens in parallel. This parallel processing significantly improves computational efficiency, enables training on massive datasets, and allows the model to capture long-range dependencies within text. These capabilities have made Transformer architectures the foundation of virtually all modern generative AI systems, including GPT, Claude, Gemini, Llama, and many other state-of-the-art language models.

A Transformer model consists of several interconnected components that work together to process language effectively. First, text is converted into numerical representations known as **embeddings**, allowing the model to represent semantic meaning mathematically. Positional encoding is then added so the model understands the order of words within a sequence. The self-attention mechanism determines which words or phrases are most relevant to one another, while multiple stacked neural network layers progressively learn increasingly complex linguistic patterns. Finally, the model predicts the most probable next token based on the context of the preceding sequence.

### Why It Matters

The Transformer architecture represents one of the most influential breakthroughs in modern artificial intelligence. Its ability to efficiently process massive amounts of textual information has enabled the development of highly capable Large Language Models that support applications such as conversational AI, document summarization, translation, code generation, knowledge retrieval, and intelligent decision support. Consequently, understanding the basic principles of the Transformer architecture provides an essential conceptual foundation for understanding how contemporary generative AI systems function.

### Example

The simplified workflow of a Transformer-based Large Language Model can be summarized as follows:

```text
User Prompt
      │
      ▼
Tokenization
      │
      ▼
Word Embeddings
      │
      ▼
Positional Encoding
      │
      ▼
Self-Attention Mechanism
      │
      ▼
Transformer Layers
      │
      ▼
Next Token Prediction
      │
      ▼
Generated Response
```

> **Key Takeaway**
>
> The Transformer architecture is the computational foundation of modern Large Language Models, enabling efficient language understanding and generation through self-attention, parallel processing, and contextual learning.

### 2.5 Prompt Engineering

Prompt engineering is the process of designing, refining, and optimizing instructions provided to generative AI systems in order to produce accurate, relevant, and contextually appropriate responses. A prompt serves as the primary interface between a user and a Large Language Model (LLM), guiding the model's reasoning, content generation, and decision-making processes. The quality, specificity, and structure of a prompt significantly influence the quality of the generated output. Consequently, effective prompt engineering has emerged as a fundamental skill for professionals who utilize generative AI in research, business, education, healthcare, software development, and numerous other domains (White et al., 2023).

Modern prompt engineering extends beyond simply asking questions. Well-designed prompts typically provide contextual information, define the desired role or perspective, specify the expected output format, establish constraints, and clearly communicate the objective of the task. By reducing ambiguity and providing sufficient context, users can substantially improve the accuracy, consistency, and usefulness of AI-generated responses. 
Prompt engineering continues to evolve alongside advances in generative AI. Contemporary techniques include zero-shot prompting, few-shot prompting, chain-of-thought prompting, role prompting, and retrieval-augmented prompting, each designed to improve model performance for specific tasks and problem domains.

### Why It Matters

Prompt engineering enables organizations to maximize the value of generative AI systems while minimizing errors, ambiguity, and inconsistent outputs. Business professionals increasingly rely on effective prompting to automate document generation, summarize complex reports, support strategic decision-making, create marketing content, generate software code, analyze customer feedback, and interact with enterprise knowledge repositories. As generative AI becomes integrated into organizational workflows, prompt engineering is emerging as a critical competency for leaders, analysts, and AI practitioners.

### Example

Consider the following prompts:

**Basic Prompt**

> Summarize Microsoft's annual report.

**Improved Prompt**

> Assume the role of a Chief Financial Officer (CFO). Summarize Microsoft's 2025 Annual Report in approximately 250 words, highlighting the company's financial performance, major business risks, strategic investments in artificial intelligence, and future growth opportunities. Present the summary using concise executive-level language.

The second prompt provides significantly more context, defines the user's role, specifies the expected output, and clearly identifies the information to be included. As a result, the generated response is typically more focused, informative, and aligned with organizational decision-making needs.

> **Key Takeaway**
>
> Prompt engineering is both an art and a scientific process that enables users to communicate effectively with Large Language Models, transforming general requests into precise instructions that produce reliable, context-aware, and actionable AI-generated outputs.

### 2.6 Tokenization

Tokenization is the process of breaking text into smaller units, known as **tokens**, that can be processed by a Large Language Model (LLM). Tokens are the fundamental building blocks of language models and may represent complete words, subwords, punctuation marks, or even individual characters, depending on the tokenization algorithm employed. Before a prompt can be interpreted by an LLM, the input text must first be converted into tokens and subsequently transformed into numerical representations that the neural network can process (Jurafsky & Martin, 2025).

Modern LLMs typically employ **subword tokenization** techniques, such as Byte Pair Encoding (BPE) or SentencePiece, to efficiently represent language while balancing vocabulary size and computational efficiency. This approach enables models to understand common words as well as unfamiliar terms, abbreviations, and domain-specific vocabulary by decomposing them into smaller, meaningful units.

The number of tokens contained within a prompt directly influences computational cost, response length, processing time, and the amount of contextual information the model can retain. Consequently, understanding tokenization enables users to design more efficient prompts, optimize interactions with LLMs, and better interpret model limitations.

### Why It Matters

Tokenization is a fundamental process underlying every interaction with a Large Language Model. Every prompt submitted to systems such as ChatGPT, Microsoft Copilot, Claude, or Gemini is first converted into tokens before the model performs language understanding or text generation. Because most commercial LLMs impose limits on the number of tokens that can be processed in a single conversation, understanding tokenization allows users to structure prompts more effectively, manage context efficiently, and reduce computational costs.

### Example

Consider the following sentence:

> **Microsoft is investing heavily in Generative AI technologies.**

A tokenizer may divide the sentence into tokens similar to the following:

| Token Number | Token |
|--------------|-------|
| 1 | Microsoft |
| 2 | is |
| 3 | investing |
| 4 | heavily |
| 5 | in |
| 6 | Generative |
| 7 | AI |
| 8 | technologies |
| 9 | . |

Although this example appears straightforward, modern tokenizers frequently divide uncommon words into multiple subword tokens to improve efficiency and enable the model to process previously unseen vocabulary.

> **Key Takeaway**
>
> Tokenization is the process of converting human language into machine-readable tokens. Every prompt and every AI-generated response begins with tokenization, making it one of the most fundamental processes underlying Large Language Models.

## 2.7 Context Window

The **context window** refers to the maximum amount of information that a Large Language Model (LLM) can process and retain during a single interaction. This information includes the user's prompt, previous conversation history, uploaded documents, system instructions, and the model's generated responses. The context window is measured in **tokens**, rather than words or characters, making tokenization a critical factor in determining how much information can be processed at one time (Bommasani et al., 2021).

As users interact with an LLM, every prompt and response contributes to the available context. When the total number of tokens exceeds the model's maximum context window, earlier portions of the conversation may be truncated or discarded to accommodate new information. Consequently, the model's ability to maintain continuity, recall previous details, and generate contextually consistent responses depends directly on the available context window. Modern LLMs vary considerably in their context window capacities. While early language models were limited to processing only a few thousand tokens, contemporary models can process hundreds of thousands or even millions of tokens, enabling the analysis of lengthy reports, research papers, legal contracts, software repositories, and enterprise knowledge bases.

### Why It Matters

Understanding the context window is essential when working with enterprise documents and knowledge-intensive applications. Business professionals frequently interact with lengthy annual reports, policy documents, financial statements, and technical documentation that may exceed the processing limits of an AI model. Knowledge of context windows enables users to structure prompts more effectively, divide large documents into manageable sections, and design AI workflows that maximize response quality while minimizing information loss.

### Example

Suppose a business analyst uploads Microsoft's Annual Report to an AI assistant and asks:

> *"Summarize the company's financial performance, AI strategy, sustainability initiatives, risk factors, and executive outlook."*

If the document exceeds the model's available context window, only a portion of the report may be processed during a single interaction. The analyst may therefore divide the report into sections or employ Retrieval-Augmented Generation (RAG) techniques to selectively retrieve the most relevant information before generating a response.

> **Key Takeaway**
>
> The context window defines how much information a Large Language Model can consider during a conversation. Understanding this limitation enables users to design more effective prompts, manage large documents efficiently, and improve the quality and consistency of AI-generated responses.

### 2.8 Multimodal Artificial Intelligence

Multimodal Artificial Intelligence (Multimodal AI) refers to AI systems capable of processing, integrating, and generating information from multiple forms of data, including text, images, audio, video, documents, and structured data. Unlike traditional AI systems that operate on a single data modality, multimodal models combine information from diverse sources to develop a richer contextual understanding of complex tasks. This capability enables AI systems to reason across different types of information, thereby improving the accuracy, flexibility, and usefulness of their responses (Bommasani et al., 2021).

Recent advances in transformer architectures have significantly accelerated the development of multimodal AI systems. Modern foundation models can simultaneously analyze written documents, interpret images, recognize speech, generate computer code, and respond to complex user queries using multiple input formats. As a result, multimodal AI is expanding the practical applications of generative AI across business, healthcare, education, engineering, finance, scientific research, and numerous other disciplines. Rather than viewing text, images, audio, and video as independent information sources, multimodal AI integrates these modalities into a unified representation, allowing models to understand relationships that would be difficult to identify using a single data type. This capability represents a major advancement toward more intelligent, context-aware AI systems that more closely resemble human communication and reasoning.

### Why It Matters

Organizations increasingly generate and manage information in multiple formats, including reports, presentations, spreadsheets, images, emails, videos, and voice recordings. Multimodal AI enables organizations to analyze these diverse information sources simultaneously, improving decision-making, knowledge discovery, customer engagement, and operational efficiency. Business leaders who understand multimodal AI will be better positioned to leverage next-generation AI technologies that support comprehensive enterprise intelligence and digital transformation.

### Example

Consider an executive reviewing Microsoft's Annual Report together with accompanying charts, financial statements, product images, and a recorded earnings call.

A multimodal AI system can:

- Summarize the written report.
- Interpret financial charts and graphs.
- Analyze product images and visual presentations.
- Transcribe and summarize executive speeches.
- Integrate insights from all these sources into a single executive briefing.

This ability to reason across multiple forms of information significantly enhances organizational decision support compared to analyzing each source independently.

> **Key Takeaway**
>
> Multimodal Artificial Intelligence extends the capabilities of Large Language Models by enabling AI systems to understand and generate information across multiple data modalities, creating richer, more context-aware interactions that support increasingly sophisticated business and organizational applications.

````python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

fig, ax = plt.subplots(figsize=(16, 11))
ax.set_xlim(0, 16)
ax.set_ylim(0, 12)
ax.axis("off")

EDGE = "#333333"

def box(x, y, w, h, color, lw=1.5, ec=EDGE):
    ax.add_patch(FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.02,rounding_size=0.10",
        linewidth=lw, facecolor=color, edgecolor=ec, zorder=2))

def label(x, y, text, fs=10, weight="bold", color="#1a1a1a", ha="center", va="center"):
    ax.text(x, y, text, ha=ha, va=va, fontsize=fs,
            fontweight=weight, color=color, zorder=4)

def pipeline_box(x, y, w, h, color, title, sub=None, title_fs=11, sub_fs=8.5):
    box(x, y, w, h, color)
    if sub:
        label(x + w/2, y + h*0.62, title, fs=title_fs)
        ax.text(x + w/2, y + h*0.28, sub, ha="center", va="center",
                fontsize=sub_fs, color="#444444", zorder=4)
    else:
        label(x + w/2, y + h/2, title, fs=title_fs)

def arrow(x1, y1, x2, y2, color=EDGE, lw=2.0, style="-|>", dashed=False):
    ax.add_patch(FancyArrowPatch(
        (x1, y1), (x2, y2),
        arrowstyle=style, mutation_scale=18,
        linewidth=lw, color=color, zorder=3,
        linestyle="--" if dashed else "-",
        shrinkA=0, shrinkB=0))

label(8, 11.55, "Figure 2.1.  The Generative AI Ecosystem and the LLM Workflow", fs=17)

# Section headers
label(2.85, 10.95, "WHAT IT IS — Conceptual Hierarchy", fs=11.5, color="#1f4e79")
label(10.6, 10.95, "HOW IT WORKS — Operational Workflow", fs=11.5, color="#5b2d82")

# Light divider between the two regions
ax.plot([5.55, 5.55], [0.9, 10.7], color="#cccccc", lw=1.2, linestyle=(0, (4, 4)), zorder=1)

# Outer -> inner (decreasing lightness reads as "zooming in")
box(0.55, 4.55, 4.6, 6.0, "#EAF2FD")          # Artificial Intelligence
box(0.95, 4.95, 3.8, 5.0, "#CFE2FA")          # Generative AI
box(1.35, 5.35, 3.0, 4.0, "#A9CCF3")          # LLMs
box(1.75, 5.75, 2.2, 2.9, "#7FB1EC")          # Transformer Architecture

label(2.85, 10.20, "Artificial Intelligence", fs=10.5, color="#10314f")
label(2.85,  9.62, "Generative AI", fs=10, color="#10314f")
label(2.85,  9.05, "Large Language Models", fs=10, color="#10314f")
label(2.85,  7.55, "Transformer\nArchitecture", fs=10, color="#0d2740")

ax.text(2.85, 4.25, "Each tier is a subset of the one around it.",
        ha="center", va="center", fontsize=8.5, style="italic", color="#555555")

CX = 10.6           # pipeline center x
W  = 5.4            # standard box width
LX = CX - W/2       # left x of standard box

# ---- Inputs (two sources that get fed into the model) ----
box(6.05, 9.45, 3.4, 1.05, "#FCE7CF")
label(7.75, 10.05, "Prompt  +  Instructions", fs=10)
ax.text(7.75, 9.72, "what you ask the model to do", ha="center", va="center",
        fontsize=8.3, color="#444444", zorder=4)

box(9.75, 9.45, 4.6, 1.05, "#FCE7CF")
label(12.05, 10.05, "Enterprise Knowledge  (RAG)", fs=10)
ax.text(12.05, 9.72, "e.g. an annual report, retrieved & added to context",
        ha="center", va="center", fontsize=8.0, color="#444444", zorder=4)

# ---- Pipeline stages ----
pipeline_box(LX, 7.95, W, 1.0, "#FFF3DE", "Tokenization", "text  →  tokens")
pipeline_box(LX, 6.55, W, 1.0, "#E3F3E5", "Embeddings", "tokens  →  numeric vectors")
pipeline_box(LX, 4.85, W, 1.25, "#EAD9F6", "Transformer Inference",
             "self-attention weighs all tokens in context")
pipeline_box(LX, 3.40, W, 1.0, "#DCC8EF", "Autoregressive Generation",
             "predicts the next token, one at a time")
pipeline_box(7.30, 1.95, 6.6, 1.05, "#D5EEF4", "Generated Output",
             "text & code (core)  ·  image / audio / video via multimodal models")
pipeline_box(7.30, 0.55, 6.6, 1.05, "#C7E7EE", "Business & Leadership Applications",
             "decision support  ·  reporting  ·  knowledge management")

# ---- Context-window bracket around the inference stage ----
cw_x = LX + W + 0.18
ax.annotate("", xy=(cw_x, 4.85), xytext=(cw_x, 6.10),
            arrowprops=dict(arrowstyle="-", color="#8a8a8a", lw=1.4), zorder=2)
ax.plot([cw_x, cw_x+0.18], [4.85, 4.85], color="#8a8a8a", lw=1.4, zorder=2)
ax.plot([cw_x, cw_x+0.18], [6.10, 6.10], color="#8a8a8a", lw=1.4, zorder=2)
ax.text(cw_x + 0.30, 5.475, "Context\nWindow\n(token budget)", ha="left", va="center",
        fontsize=8.0, color="#666666", zorder=4)

# inputs -> tokenization
arrow(7.75, 9.45, CX - 0.9, 8.95)
arrow(12.05, 9.45, CX + 0.9, 8.95)
# main pipeline
arrow(CX, 7.95, CX, 7.55)            # tokenization -> embeddings
arrow(CX, 6.55, CX, 6.10)            # embeddings -> inference
arrow(CX, 4.85, CX, 4.40)            # inference -> generation
arrow(CX, 3.40, CX, 3.00)            # generation -> output
arrow(CX, 1.95, CX, 1.60)            # output -> applications

arrow(3.95, 7.05, LX - 0.05, 5.55, color="#5b2d82", lw=1.8, dashed=True)
ax.text(6.05, 6.55, "powers", fontsize=8.5, style="italic",
        color="#5b2d82", rotation=-30, zorder=4)
arrow(3.95, 7.05, LX - 0.05, 5.55, color="#5b2d82", lw=1.8, dashed=True)
ax.text(6.05, 6.55, "powers", fontsize=8.5, style="italic",
        color="#5b2d82", rotation=-30, zorder=4)

plt.savefig("images/Figure_2_1_Generative_AI_Ecosystem.png", dpi=300, bbox_inches="tight")
plt.show()
````

Unstructured text refers to information that does not follow a predefined database schema. Annual reports, emails, contracts, technical documentation, meeting minutes, and organizational policies are all examples of unstructured data. Although these documents contain valuable knowledge, they require preprocessing before they can be used by machine learning models and LLMs.

### 3.6 Prompt Engineering Demonstration

Prompt engineering is the process of designing clear and effective instructions that guide a Large Language Model (LLM) toward producing accurate, relevant, and useful responses. The quality of an LLM's output depends not only on the underlying model but also on the quality of the prompt provided by the user.

Enterprise AI applications frequently use organizational documents such as annual reports, policy manuals, contracts, and technical documentation as contextual information. Rather than asking vague questions, users provide well-structured prompts that specify the task, the desired audience, the expected format, and any relevant business context.

In this demonstration, we will examine several prompts based on the Microsoft Annual Report. Although we are not yet sending these prompts to an LLM, the examples illustrate how prompt quality influences the effectiveness of AI-assisted decision making.

````python
# Example prompts based on the Microsoft Annual Report

basic_prompt = """
Summarize the Microsoft Annual Report.
"""

improved_prompt = """
You are a senior business analyst.

Using the Microsoft Annual Report, summarize Microsoft's strategic priorities,
major financial achievements, and artificial intelligence initiatives.

Present the summary as five concise bullet points suitable for executive leadership.
"""

advanced_prompt = """
Assume you are advising the Chief Executive Officer.

Analyze the Microsoft Annual Report and identify:

1. Three strategic priorities.
2. Three business risks.
3. Three opportunities created by Generative AI.
4. Recommendations for executive decision making.

Present your findings using clear headings and concise business language.
"""

print("Basic Prompt")
print("-" * 60)
print(basic_prompt)

print("\nImproved Prompt")
print("-" * 60)
print(improved_prompt)

print("\nAdvanced Prompt")
print("-" * 60)
print(advanced_prompt)
````

#### Discussion

These examples demonstrate that prompt engineering extends beyond simply asking a question. Effective prompts provide context, define the user's role, specify the required task, identify the intended audience, and describe the desired output format.

As prompts become more detailed and structured, Large Language Models are better able to generate responses that align with organizational objectives and support informed decision making. Throughout this course, prompt engineering will become an essential skill for interacting effectively with enterprise AI systems.

### 3.7 Tokenization Demonstration

Although humans read documents as words, sentences, and paragraphs, Large Language Models process text as **tokens**. A token is the basic unit of text that an LLM uses during training and inference. Depending on the tokenizer, a token may represent an entire word, part of a word, punctuation, or even whitespace.

Tokenization is the first computational step performed before a prompt reaches the transformer architecture. Every prompt and every generated response is converted into a sequence of tokens that the model can process numerically.

In this demonstration, we illustrate the concept of tokenization using a simple whitespace tokenizer. Although modern LLMs employ more sophisticated tokenization algorithms, this example provides an intuitive understanding of how textual information is transformed into smaller computational units.

````python
sample_text = """
Microsoft is investing heavily in Artificial Intelligence to improve
productivity, cloud computing, cybersecurity, and enterprise innovation.
"""

print("Original Text")
print("-" * 60)
print(sample_text)

tokens = sample_text.split()

print("\nTokenized Text")
print("-" * 60)

for i, token in enumerate(tokens, start=1):
    print(f"{i:2}. {token}")

print("\nTotal Tokens:", len(tokens))
````

````python
sample_paragraph = paragraphs[10]

print("Enterprise Text Sample")
print("-" * 60)
print(sample_paragraph)

tokens = sample_paragraph.split()

print("\nNumber of Words (Approximate Tokens)")
print("-" * 60)
print(len(tokens))
````

#### Discussion

This demonstration uses a simple whitespace tokenizer to illustrate the concept of tokenization. Modern Large Language Models employ specialized tokenization algorithms such as Byte Pair Encoding (BPE), WordPiece, or SentencePiece, which often divide words into smaller subword units rather than separating text solely by spaces.

Although the exact number of tokens produced by an LLM tokenizer differs from this simplified example, the underlying concept remains the same: textual information must first be converted into a sequence of tokens before it can be processed by the transformer architecture. Understanding tokenization is essential because model context windows, computational cost, inference speed, and API pricing are typically measured in tokens rather than words or pages.
